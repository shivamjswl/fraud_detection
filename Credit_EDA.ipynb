### Checking the imbalance of the data set by using target variable

Target variable in this data set is "TARGET"

If the value in "TARGET" column is "1"

then it means client in that row has payment difficulties like he/she had late payment more than X days on at least one of the first Y installments of the loan in our sample.

If the value in "TARGET" column is "0"

then it means All other cases when the payment is done on time by client in that row.

```python
# Checking the imbalance.

inp0["TARGET"].value_counts(normalize=True)*100
```

```python
# Checking the imbalance by viewing pie chart.

plt.pie(inp0["TARGET"].value_counts(normalize=True)*100,labels=['NON-DEFAULT (TARGET=0)','DEFAULT (TARGET=1)'],explode=(0,0.05),autopct='%.2f%%')
plt.title('TARGET Variable - DEFAULTER Vs NON-DEFAULTER')
plt.show()
```
